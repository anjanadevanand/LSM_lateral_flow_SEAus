### Get model data from the streamflow text file

```python
##############################################
# Get lat-lon of gauges from fulldom_hires
##############################################

ds_rt = xr.open_dataset(rt_stack_dir + fname)

RT_resol = 250

gauge_list = df_hrs['FID'].values
lat_gauge_list = []
lon_gauge_list = []
flowacc_gauge_list = []

for gauge in gauge_list:
    # print(gauge)
    lat_arr = ds_rt['LATITUDE'].where(ds_rt['frxst_pts']==gauge).values
    # print('latitude = '  + str(lat_arr[~np.isnan(lat_arr)]))
    lat_gauge_list.append(lat_arr[~np.isnan(lat_arr)][0])
    lon_arr = ds_rt['LONGITUDE'].where(ds_rt['frxst_pts']==gauge).values
    # print('longitude = ' + str(lon_arr[~np.isnan(lon_arr)]))
    lon_gauge_list.append(lon_arr[~np.isnan(lon_arr)][0])
    flowacc_arr = ds_rt['FLOWACC'].where(ds_rt['frxst_pts']==gauge).values*RT_resol*RT_resol/1000000
    # print('flow accum = ' + str(flowacc_arr[~np.isnan(flowacc_arr)]))
    flowacc_gauge_list.append(flowacc_arr[~np.isnan(flowacc_arr)][0])
```
