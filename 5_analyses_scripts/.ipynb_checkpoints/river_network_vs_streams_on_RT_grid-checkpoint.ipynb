```python
# just get a temporary 
#wrf_hydro_temp = '/g/data/w28/ad9701/WRF-Hydro/SEA/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_LSMonly_defSoil_defPar/trunk/NDHMS/Run/'
#wrf_hydro_temp = '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/'
#ds_temp = xr.open_dataset(wrf_hydro_temp + 'OUTPUT/DAILY_FILES/LH_concatTime.nc')

# temp_file = '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_t80_RTCHon_12g_monPnTCorr_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/RTOUT_6hr/20171231.RTOUT_DOMAIN1'
# ds_temp = xr.open_dataset(temp_file)
# X, Y = np.meshgrid(ds_temp.x.values, ds_temp.y.values)
# source = Proj(ds_temp.proj4)

da_ch_grid = xr.where(ds_fulldom['CHANNELGRID'] > -2, 1, np.nan)
da_ch_grid_so_4 = da_ch_grid.where(ds_fulldom['STREAMORDER'] > 4)
da_blank = xr.where(ds_fulldom['CHANNELGRID'] > 5, 1, np.nan)

X, Y = np.meshgrid(ds_fulldom.x.values, ds_fulldom.y.values)
source = Proj(ds_fulldom.proj4)
lon, lat = source(X.flatten(), Y.flatten(), inverse=True)
lon_2d = lon.reshape(X.shape)
lat_2d = lat.reshape(Y.shape)
da_ch_grid_so_4.coords['lat'] = (da_ch_grid_so_4.dims, lat_2d)
da_ch_grid_so_4.coords['lon'] = (da_ch_grid_so_4.dims, lon_2d)

da_blank.coords['lat'] = (da_blank.dims, lat_2d)
da_blank.coords['lon'] = (da_blank.dims, lon_2d)


da_ch_grid = xr.where(ds_fulldom_100m['CHANNELGRID'] > -2, 1, np.nan)
da_ch_grid_so_4_100m = da_ch_grid.where(ds_fulldom_100m['STREAMORDER'] > 4)

X, Y = np.meshgrid(ds_fulldom_100m.x.values, ds_fulldom_100m.y.values)
source = Proj(ds_fulldom_100m.proj4)
lon, lat = source(X.flatten(), Y.flatten(), inverse=True)
lon_2d = lon.reshape(X.shape)
lat_2d = lat.reshape(Y.shape)
da_ch_grid_so_4_100m.coords['lat'] = (da_ch_grid_so_4_100m.dims, lat_2d)
da_ch_grid_so_4_100m.coords['lon'] = (da_ch_grid_so_4_100m.dims, lon_2d)



xticks = [144, 146, 148]
yticks = [-34, -36, -38]

projection = ccrs.LambertConformal(central_latitude=-35.7, central_longitude=147.12, standard_parallels=(-35.7, -35.7), cutoff=0)


sh1 = '/g/data/w97/ad9701/WRF-Hydro/SEA/shapefiles/rbasin_polygon_SEA_subset.shp'
sh2 = '/g/data/w97/ad9701/WRF-Hydro/SEA/shapefiles/rivers_so4_and_higher.shp'

# to plot SW basin boundaries & rivers
shapefile = [sh1, sh2]
shapefile_color=['darkgrey', 'deeppink']
sh_lwd = [0.5, 2]
sh_alpha = 0.3

# to plot  rivers
shapefile = [sh2]
shapefile_color=['deeppink']
sh_lwd = [1]
sh_alpha = 0.3
```
