### Saved domain average Tmax data in netcdf files using the below code

```python
mdlVar = 'T2M'
for dir_list in [wrf_hydro_ctl_list, wrf_hydro_lat_list, wrf_hydro_lat_list_250]:
    for mdl_dir in dir_list:
        ds_mdl = xr.open_dataset(mdl_dir + 'T2M_daily_max.nc')
        da_mdl = ds_mdl['T2M']
        da_mdl_mean = da_mdl.resample(time = 'M').mean()
        da_mdl_max = da_mdl.resample(time = 'M').max()
        da_mdl_mean_domAvg = da_mdl_mean.mean(['x', 'y'])
        da_mdl_max_domAvg = da_mdl_max.mean(['x', 'y'])
        # save the data in a netcdf file & read it in later

        out_dir = mdl_dir + 'domainAvg/'
        if not os.path.exists(out_dir):
            os.makedirs(out_dir)
        out_file = out_dir + mdlVar + '_mean_monthly.nc'
        print(out_file)
        da_mdl_mean_domAvg.to_netcdf(out_file)
        out_file = out_dir + mdlVar + '_max_monthly.nc'
        print(out_file)
        da_mdl_max_domAvg.to_netcdf(out_file)
```
