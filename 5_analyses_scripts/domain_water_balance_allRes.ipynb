```python
import xarray as xr
import numpy as np
import pandas as pd
import os
import sys
import glob
import xesmf as xe
from pyproj import Proj
import cartopy.crs as ccrs

new_path = '/home/566/ad9701/drought_probability/final_code_AU/'
if new_path not in sys.path:
    
    sys.path.append(new_path)
import validation_functions as myfuncs

import matplotlib.pyplot as plt

SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 16

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

def add_latlon_coords(da):
    if da is not None:
        X, Y = np.meshgrid(da.x.values, da.y.values)
        source = Proj(da.attrs['esri_pe_string'])
        lon, lat = source(X.flatten(), Y.flatten(), inverse=True)
        lon_2d = lon.reshape(X.shape)
        lat_2d = lat.reshape(Y.shape)
    
        da.coords['lat'] = (da.dims, lat_2d)
        da.coords['lon'] = (da.dims, lon_2d)
        return(da)
    else:
        return None

def calc_tot_sm_inmm(da_vol_sm, soil_depth_mm = [100, 300, 600, 1000]):
    da_tot_sm = da_vol_sm.isel(soil_layers_stag=0)*soil_depth_mm[0] + \
    da_vol_sm.isel(soil_layers_stag=1)*soil_depth_mm[1] + \
    da_vol_sm.isel(soil_layers_stag=2)*soil_depth_mm[2] + \
    da_vol_sm.isel(soil_layers_stag=3)*soil_depth_mm[3]
    return(da_tot_sm)

# Projection information for the domain

#********** Domain 4 ********************
standard_parallels = (-35.7, -35.7)
central_lat = -35.7
central_lon = 147.12
#****************************************

#********** Domain 3 ********************
# standard_parallels = (-36.854, -36.854)
# central_lat = -36.854
# central_lon = 147.078
#****************************************
```

```python
# function to calculate water budgets in mm
def calc_water_budget(input_dir, startTime, endTime, 
                      varnames = ['Precip.', 'ET', 'Soil Mois.', 'sfcrnoff', 'udgrnoff'],  #, 'boundary', 'sfcheadsubrt'], # removed extra vars as I'm using them in the final calc.
                      filenames = ['ACCPRCP_concatTime.nc', 'ACCET_concatTime.nc','SOIL_M_concatTime_*.nc', 'SFCRNOFF_concatTime.nc', 'UGDRNOFF_concatTime.nc'], #, None, None], 
                      RT = False):
    '''
    Input directory containing the files as well as the RTOUT files if required
    startTime and end Time are strings of form YYYYMMDDTHH
    '''
    dict_filenames = dict(zip(varnames, filenames))
    
    da_budget_dict = {}
    for var in varnames:
        if ((var == 'boundary') | (var == 'sfcheadsubrt')) & (RT is False):
            da_budget_dict.update({var: None})
        else:  
            if ((var == 'sfcrnoff') | (var == 'boundary') | (var == 'sfcheadsubrt')) & (RT is True):
                # in my new set of runs the RT files may be saved in one of 3 different places, depending upon the stage of processing/archival of the run
                # 1. in the input_dir, 2. in input_dir/RTOUT_6hr/, and 3. input_dir/RTOUT_instances/
                # check each of these locations sequentially
                f_start = glob.glob(input_dir + startTime[0: 8] + '.RTOUT_DOMAIN1')
                f_end = glob.glob(input_dir + endTime[0: 8] + '.RTOUT_DOMAIN1')
                if len(f_start) < 1:
                    f_start = glob.glob(input_dir + '/RTOUT_6hr/' + startTime[0: 8] + '.RTOUT_DOMAIN1')
                    f_end = glob.glob(input_dir + '/RTOUT_6hr/' + endTime[0: 8] + '.RTOUT_DOMAIN1')
                    if len(f_start) < 1:
                        f_start = glob.glob(input_dir + '/RTOUT_instances/' + startTime[0: 8] + '.RTOUT_DOMAIN1')
                        f_end = glob.glob(input_dir + '/RTOUT_instances/' + endTime[0: 8] + '.RTOUT_DOMAIN1')
                if (var == 'sfcrnoff') | (var == 'sfcheadsubrt'):
                    if var == 'sfcrnoff':
                        # use the RT files to get the surface runoff
                        var_name = 'QSTRMVOLRT'
                    if var == 'sfcheadsubrt':
                        var_name = 'sfcheadsubrt'
                    ds1 = xr.open_dataset(f_start[0])
                    ds2 = xr.open_dataset(f_end[0])
                    da_start = ds1[var_name].sel(time = startTime).drop('reference_time', errors = 'ignore')
                    da_end = ds2[var_name].sel(time = endTime).drop('reference_time', errors = 'ignore')
                    da_diff =  da_end - da_start
                    esri_pe_string = ds1[var_name].attrs['esri_pe_string']
                if var == 'boundary':
                    bdy_name = 'QBDRYRT'
                    ds1 = xr.open_dataset(f_start[0])
                    ds2 = xr.open_dataset(f_end[0])
                    da_start = ds1[bdy_name].sel(time = startTime).drop('reference_time', errors = 'ignore')
                    da_start = da_start*-1
                    da_end = ds2[bdy_name].sel(time = endTime).drop('reference_time', errors = 'ignore')
                    da_end = da_end*-1
                    da_diff =  da_end - da_start
                    esri_pe_string = ds1[bdy_name].attrs['esri_pe_string']
            else:
                if '*' in dict_filenames[var]:
                    ds = xr.open_mfdataset(input_dir + dict_filenames[var])
                else:
                    ds = xr.open_dataset(input_dir + dict_filenames[var])
                file_vname = list(ds.keys())[0]
                da_start = ds[file_vname].sel(time = startTime)
                da_end = ds[file_vname].sel(time = endTime)
                if var == 'Soil Mois.':
                    da_start_tot = calc_tot_sm_inmm(da_start)
                    da_end_tot = calc_tot_sm_inmm(da_end)
                    da_diff = da_end_tot - da_start_tot
                else:
                    da_diff = da_end - da_start
                esri_pe_string = da_start.attrs['esri_pe_string']
    
            da_diff = da_diff.assign_attrs({'units':'mm', 
                                            'esri_pe_string': esri_pe_string,
                                            'description': var + ' gain',
                                            'startTime': startTime,
                                            'endTime': endTime,
                                            'data_directory': input_dir})
            da_budget_dict.update({var: da_diff})
    return(da_budget_dict)
```

```python
# function to calculate water budgets in mm
def calc_domAvg_accum_water_budget(input_dir, startTime, endTime, 
                                   varnames = ['Precip.', 'ET', 'Soil Mois.'],
                                   filenames = ['ACCPRCP_concatTime.nc', 'ACCET_concatTime.nc', 'SOIL_M_total_inmm_00hr_*.nc']):
    '''
    Input directory containing the files as well as the RTOUT files if required
    startTime and endTime are strings of form YYYYMMDD
    '''
    dict_filenames = dict(zip(varnames, filenames))
    
    da_domAvg_budget_dict = {}
    for var in varnames:
        if '*' in dict_filenames[var]:
            ds = xr.open_mfdataset(input_dir + dict_filenames[var])
        else:
            ds = xr.open_dataset(input_dir + dict_filenames[var])
        file_vname = list(ds.keys())[0]
        da_subset = ds[file_vname].sel(time = slice(startTime, endTime))
        # subset time to get only the 00 hr
        target_hours = [0]
        da_subset = da_subset.sel(time=da_subset['time.hour'].isin(target_hours))
        if var == 'Precip.':
            # da_wlatlon = add_latlon_coords(da_subset[0,:,:])
            da_singleTime = da_subset[0,:,:]
            # create a mask from the Precip data. Works as precip is the first variable.
            da_mask = xr.where(np.isnan(da_singleTime), np.nan, 1).rename('land_mask')
        # now calculate domain averages; masking as necessary
        da_subset_domAvg = (da_subset*da_mask).mean(['x', 'y']).load()
    
        # get the mm change for each day 
        da_subset_domAvg = da_subset_domAvg.diff('time')
        # now get the cumulative moisture in that component of the budget
        da_subset_domAvg = da_subset_domAvg.cumsum('time')
        
        da_subset_domAvg = da_subset_domAvg.assign_attrs({'units':'mm', 
                                                          'description': var + ' cumulative',
                                                          'startTime': startTime,
                                                          'endTime': endTime,
                                                          'data_directory': input_dir})
        da_domAvg_budget_dict.update({var: da_subset_domAvg})
    return(da_domAvg_budget_dict)
```

```python
startTime_list = ['20151201T00']
endTime_list = ['20171201T00']

overall_ctl_budgets_domAvg_list = []
overall_lat_budgets_domAvg_list = []
overall_lat_250_budgets_domAvg_list = []

for startTime, endTime in zip(startTime_list, endTime_list):
    temp_ctl_budgets_list = []
    for ctl_dir in wrf_hydro_ctl_list:
        CTL_budget_dict = calc_water_budget(ctl_dir, startTime = startTime, endTime = endTime, RT = False)
        # use one consistent land mask for CTL & LAT runs
        da_wlatlon = add_latlon_coords(CTL_budget_dict['Precip.'])
        da_mask = xr.where(da_wlatlon==0, np.nan, 1).rename('land_mask')
        # add lat-lon co-ords
        for var in list(CTL_budget_dict.keys()):
            da_wlatlon = add_latlon_coords(CTL_budget_dict[var])
            # if var == 'Precip.':
            #     da_mask = xr.where(np.isnan(da_wlatlon), np.nan, 1).rename('land_mask')
            CTL_budget_dict.update({var: da_wlatlon})
            # now calculate domain averages; masking as necessary
            ctl_var_avg = (da_wlatlon*da_mask).mean(['x', 'y']).values
            # *** if looking for spatial plots - comment the line below & remove calc of domain averages
            CTL_budget_dict.update({var: ctl_var_avg})
        for d in [CTL_budget_dict]:
            d['Runoff'] = d['Precip.'] - d['ET'] - d['Soil Mois.']
        temp_ctl_budgets_list.append(CTL_budget_dict)
    overall_ctl_budgets_domAvg_list.append(temp_ctl_budgets_list)
        
    temp_lat_budgets_list = []
    for lat_dir in wrf_hydro_lat_list:
        LAT_budget_dict = calc_water_budget(lat_dir, startTime = startTime, endTime = endTime, RT = True)
        for var in list(LAT_budget_dict.keys()):
            da_wlatlon = add_latlon_coords(LAT_budget_dict[var])
            # if var == 'Precip.':
            #     da_mask = xr.where(np.isnan(da_wlatlon), np.nan, 1).rename('land_mask')
            LAT_budget_dict.update({var: da_wlatlon})
            if var == 'sfcrnoff':
                lat_var_avg = da_wlatlon.mean(['x', 'y']).values
            else:
                lat_var_avg = (da_wlatlon*da_mask).mean(['x', 'y']).values
            LAT_budget_dict.update({var: lat_var_avg})
        for d in [LAT_budget_dict]:
            d['Runoff'] = d['Precip.'] - d['ET'] - d['Soil Mois.']
        temp_lat_budgets_list.append(LAT_budget_dict)
    overall_lat_budgets_domAvg_list.append(temp_lat_budgets_list)
    
    temp_lat_250_budgets_list = []
    for lat_dir in wrf_hydro_lat_list_250:
        LAT_budget_dict = calc_water_budget(lat_dir, startTime = startTime, endTime = endTime, RT = True)
        for var in list(LAT_budget_dict.keys()):
            da_wlatlon = add_latlon_coords(LAT_budget_dict[var])
            # if var == 'Precip.':
            #     da_mask = xr.where(np.isnan(da_wlatlon), np.nan, 1).rename('land_mask')
            LAT_budget_dict.update({var: da_wlatlon})
            if var == 'sfcrnoff':
                lat_var_avg = da_wlatlon.mean(['x', 'y']).values
            else:
                lat_var_avg = (da_wlatlon*da_mask).mean(['x', 'y']).values
            LAT_budget_dict.update({var: lat_var_avg})
        for d in [LAT_budget_dict]:
            d['Runoff'] = d['Precip.'] - d['ET'] - d['Soil Mois.']
        temp_lat_250_budgets_list.append(LAT_budget_dict)
    overall_lat_250_budgets_domAvg_list.append(temp_lat_250_budgets_list)
```
