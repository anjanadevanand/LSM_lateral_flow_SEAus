```python
fig_dir = '/home/566/ad9701/figures/wrf_hydro_standalone/finalRuns_allRes_noBaseFlow/subbasin_avg/'

########################
# FILENAMESa
########################
# the filenames
# fname_list = ['RAINRATE', 'SOIL_M_2layer', 'LH']
# fname_suffix_list = ['_daily_mean']*3

fname_list = ['RAINRATE', 'LH', 'T2M']
fname_suffix_list = ['_daily_mean', '_daily_mean', '_daily_max']

########################
# VARNAMES
########################                         
# varname_list = ['RAINRATE', 'SOIL_M', 'LH']
varname_list = ['RAINRATE', 'LH', 'T2M']

######################################
# GET THE SUB-BASIN MASKS FOR SUBSET
######################################

basin_mask_dir = '/g/data/w97/ad9701/WRF-Hydro/SEA/shapefiles/individual_basins/'
#basin_list = ['upper_basins', 'ovens', 'goulburn', 'broken', 'murray_riverina']
basin_list = ['upper_basins', 'ovens', 'murray_riverina']

basin_mask_list_4k = []
for basin in basin_list:
    ds_mask = xr.open_dataset(basin_mask_dir + basin + '_mask.nc')
    basin_mask_list_4k.append(ds_mask['mask'])

basin_mask_list_1k = []
for basin in basin_list:
    ds_mask = xr.open_dataset(basin_mask_dir + basin + '_mask_1km.nc')
    basin_mask_list_1k.append(ds_mask['mask'])

basin_mask_list_10k = []
for basin in basin_list:
    ds_mask = xr.open_dataset(basin_mask_dir + basin + '_mask_10km.nc')
    basin_mask_list_10k.append(ds_mask['mask'])

#time_sel = None
#time_sel = slice('2015-12', '2017-12')
time_sel = slice('2016', '2017')


wrf_hydro_lat_list = [simpath.lat_4km_250rt, simpath.lat_4km_100rt,
                      simpath.lat_1km_250rt, simpath.lat_1km_100rt,
                      simpath.lat_10km_250rt, simpath.lat_10km_100rt]
wrf_hydro_lat_list = [x + '/OUTPUT/DAILY_FILES/' for x in wrf_hydro_lat_list]

wrf_hydro_ctl_list = [simpath.ctl_4km, simpath.ctl_4km,
                      simpath.ctl_1km, simpath.ctl_1km,
                      simpath.ctl_10km, simpath.ctl_10km]
wrf_hydro_ctl_list = [x + '/OUTPUT/DAILY_FILES/' for x in wrf_hydro_ctl_list]

run_name_list = ['LAT4-250', 'LAT4-100',
                 'LAT1-250', 'LAT1-100',
                 'LAT10-250', 'LAT10-100']


# # wrf_hydro_ctl = '/g/data/w28/ad9701/WRF-Hydro/SEA/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'
# # wrf_hydro_lat = '/g/data/w28/ad9701/WRF-Hydro/SEA/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_12g_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'
# # run_name = '4km'

# #**** The selected run without baseflow ***
# #### 4-km ####
# wrf_hydro_lat_list = ['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_t80_RTCHon_12g_monPnTCorr_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
#                       '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km_100mRT/wrf_hydro_nwm_public-5.2.0_domain4_4km_t500_RTCHon_4g_monPnTCorr_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/']
# wrf_hydro_ctl_list = ['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',      ### CC*
#                       '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_4km/wrf_hydro_nwm_public-5.2.0_domain4_4km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/']
# #run_name_list = ['defSoil_defPar_4km', 'defSoil_defPar_4km_100RT']
# run_name_list = ['LAT4-250', 'LAT4-100']

# #### 1-km ####
# wrf_hydro_ctl_list.extend(['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_1km/wrf_hydro_nwm_public-5.2.0_domain4_1km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
#                            '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_1km/wrf_hydro_nwm_public-5.2.0_domain4_1km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'])   ### CC
# wrf_hydro_lat_list.extend(['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_1km/wrf_hydro_nwm_public-5.2.0_domain4_1km_monPnTCorr_t80_RTCHon_monPnTCorr_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
#                            '/scratch/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_1km_100mRT/output_wrf_hydro_nwm_public-5.2.0_domain4_1km_monPnTCorr_t500_RTCHon_monPnTCorr_defSoil_defPar/OUTPUT/DAILY_FILES/'])
# run_name_list.extend(['LAT1-250', 'LAT1-100'])

# #### 10-km ####

# wrf_hydro_lat_list.extend(['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_t80_RTCHon_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
#                            '/g/data/w28/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_10km_167mRT/wrf_hydro_nwm_public-5.2.0_domain4_10km_t180_RTCHon_monPnTCorr_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'])

# wrf_hydro_ctl_list.extend(['/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
#                            '/g/data/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'])


# # wrf_hydro_lat_list.extend(['/scratch/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wNewCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_t80_RTCHon_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'])
# #                       #*** 100m RT run in progress ***
# # wrf_hydro_ctl_list.extend(['/scratch/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wNewCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/',
# #                            '/scratch/w97/ad9701/WRF-Hydro/SEA/noGW_runs/runs_wNewCalibPars_10km/wrf_hydro_nwm_public-5.2.0_domain4_10km_monPnTCorr_LSMonly_defSoil_defPar/trunk/NDHMS/Run/OUTPUT/DAILY_FILES/'])

# run_name_list.extend(['LAT10-250', 'LAT10-100'])

basin_mask_list_byRes = [basin_mask_list_4k, basin_mask_list_4k, basin_mask_list_1k, basin_mask_list_1k, basin_mask_list_10k, basin_mask_list_10k]

def calc_basin_avg(da_var, da_mask):
    da_var_masked = da_var*da_mask
    da_var_masked_avg = da_var_masked.mean(['x', 'y'])
    return da_var_masked_avg

lat_allRuns = {}
ctl_allRuns = {}
for wrf_hydro_ctl, wrf_hydro_lat, run_name, basin_mask_list in zip(wrf_hydro_ctl_list, wrf_hydro_lat_list, run_name_list, basin_mask_list_byRes):
    # save basin averages for all variables
    lat_dict_basin_averages = {}
    ctl_dict_basin_averages = {}
    for fname, suffix, varname in zip(fname_list, fname_suffix_list, varname_list):
        ds_ctl = xr.open_dataset(wrf_hydro_ctl + fname + suffix + '.nc')
        ds_lat = xr.open_dataset(wrf_hydro_lat + fname + suffix + '.nc')
        da_ctl = ds_ctl[varname]
        da_lat = ds_lat[varname]
        if time_sel is not None:
            da_ctl = da_ctl.sel(time = time_sel)
            da_lat = da_lat.sel(time = time_sel)
        if varname == 'RAINRATE':
            # convert it to mm/day
            attrs = da_ctl.attrs
            attrs.update({'long_name':'Precipitation', 'units':'mm/day'})
            da_ctl = (da_ctl*86400).assign_attrs(attrs)
            attrs = da_lat.attrs
            attrs.update({'long_name':'Precipitation', 'units':'mm/day'})
            da_lat = (da_lat*86400).assign_attrs(attrs)
        ctl_var_in_basin = []
        lat_var_in_basin = []
        for basin, da_mask in zip(basin_list, basin_mask_list):
            da_ctl_basin_avg = calc_basin_avg(da_ctl, da_mask)
            da_lat_basin_avg = calc_basin_avg(da_lat, da_mask)
            ctl_var_in_basin.append(da_ctl_basin_avg)
            lat_var_in_basin.append(da_lat_basin_avg)
    
        # save the basin averages into a dictionary
        lat_dict_basin_averages.update({fname :lat_var_in_basin})
        ctl_dict_basin_averages.update({fname :ctl_var_in_basin})
    lat_allRuns.update({run_name: lat_dict_basin_averages})
    ctl_allRuns.update({run_name: ctl_dict_basin_averages})
```
